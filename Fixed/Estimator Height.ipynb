```python
import pandas as pd
import os

import numpy as np
import math
import itertools
import random
from sklearn.externals import joblib

import matplotlib.pyplot as plt
import math
```

```python
label_file = "./height.csv"
```

```python
data_df = pd.read_csv(label_file)
data_df.drop(['Unnamed: 0'], axis=1)
```

```python
y_height = data_df.height.values ## all labels
y_distance = data_df.length.values
```

```python
X_train = []
```

```python
for i in  y_distance:
    X_train.append([0,i])
```

```python
X_train
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_height_train, y_height_test = train_test_split(X_train, y_height, test_size=0.2, random_state=1)
```

```python
from sklearn.kernel_ridge import KernelRidge
from sklearn import  linear_model
from sklearn.svm import SVR
from sklearn.linear_model import Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from math import sqrt
```

```python
y_height_estimation = []
y_height_true = []
```

```python
def report_goodness(model,X_test,y_test,predictor_log=True):
    # Make predictions using the testing set
    y_pred = model.predict(X_test)
    y_true = y_test
    y_height_estimation = y_pred
    if predictor_log:
        y_height_true.append(y_true)
        y_true = np.log(y_test)
    print('Model Performance')
    print("Root Mean squared error: %.4f" %sqrt(mean_squared_error(y_true, y_pred)))
    print("Mean squared error: %.4f" %mean_squared_error(y_true, y_pred))
    print('Mean Absolute Error: %.4f' %mean_absolute_error(y_true, y_pred))
    print('Variant score: %.4f' %r2_score(y_true, y_pred))
```

```python
model_linear = linear_model.LinearRegression()
model_linear = model_linear.fit(X_train,np.log(y_height_train))

report_goodness(model_linear,X_test,y_height_test)
```

```python
model_lasso = Lasso(fit_intercept=True, alpha=0.0015, random_state=4, normalize=True)
model_lasso = model_lasso.fit(X_train,np.log(y_height_train))

report_goodness(model_lasso,X_test,y_height_test)
```

```python
model_ridge = Ridge(fit_intercept=True, alpha=0.0015, random_state=4, normalize=True)
model_ridge = model_ridge.fit(X_train,np.log(y_height_train))

report_goodness(model_ridge,X_test,y_height_test)
```

```python
height_model = 'height_predictor_height_best.model'
joblib.dump(model_ridge, height_model)
```
